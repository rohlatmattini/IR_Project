## 📊 Step 8: Final Summary Table — All Models Comparison


```python
results_data = []

for model_name, before, after in [
    ("BM25", bm25_before, bm25_after),
    ("TF-IDF", tfidf_before, tfidf_after),
    ("Embedding", emb_before, emb_after),
    ("Hybrid_Parallel", hp_before, hp_after),
    ("Hybrid_Serial", hs_before, hs_after),
]:
    if before and after:
        results_data.append({
            "Model": model_name,
            "Queries": before['num_queries_evaluated'],
            "precision@10 (Before)": round(before['P@10'], 4),
            "precision@10 (After)": round(after['P@10'], 4),
            "recall@10 (Before)": round(before['Recall'], 4),
            "recall@10 (After)": round(after['Recall'], 4),
            "MAP@10 (Before)": round(before['MAP'], 4),
            "MAP@10 (After)": round(after['MAP'], 4),
            "nDCG@10 (Before)": round(before['nDCG@10'], 4),
            "nDCG@10 (After)": round(after['nDCG@10'], 4),
        })

df = pd.DataFrame(results_data)

print("="*100)
print(" 📊 FINAL EVALUATION SUMMARY")
print(f" Total Queries: 49 (ALL FROM QRELS)")
print("="*100)
print(df.to_string(index=False))

df.to_csv("evaluation_results.csv", index=False)
print(f"\n💾 Saved to: evaluation_results.csv")

df
```
